```python
#------------------------------------------
#　#3.正解率を計算する関数
#------------------------------------------
def calculate_accuracy(predicted_data, true_data):
    total_reviews = len(predicted_data)
    correct_high_predictions = 0
    correct_low_predictions = 0

    for pred, true in zip(predicted_data, true_data):
        # 高評価要素の一致を確認し、一致した場合（正解の場合）はカウント
        if set(pred['高評価要素']) == set(true['高評価要素']):
            correct_high_predictions += 1
        # 低評価要素の一致を確認、一致した場合（正解の場合）はカウント
        if set(pred['低評価要素']) == set(true['低評価要素']):
            correct_low_predictions += 1

    # 高評価要素と低評価要素の正解率の計算
    highly_rated_accuracy = correct_high_predictions / total_reviews
    low_rated_accuracy = correct_low_predictions / total_reviews

    return highly_rated_accuracy, low_rated_accuracy
```
